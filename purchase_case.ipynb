Label Encoding is a technique used to convert categorical variables into numerical format, making it easier for machine learning models to process them. It assigns a unique integer to each category in the data.

```python
from sklearn.preprocessing import LabelEncoder

# Sample data
colors = ["Red", "Blue", "Green", "Blue", "Red"]

# Initialize encoder
encoder = LabelEncoder()

# Fit and transform
encoded_colors = encoder.fit_transform(colors)

print(encoded_colors)
```

```python
from sklearn.preprocessing import LabelEncoder

# Dictionary to store mappings

label_mappings = {}

# Iterate over columns as apply label encoder

for i in df.columns:
  if df[i].dtype == 'object': # checking if the column is categorical
    le = LabelEncoder()
    df[i] = le.fit_transform(df[i]) # encoded the column
    label_mappings[i] = dict(zip(le.classes_,le.transform(le.classes_)))

# le.classes_ - it will contain original unique categories (male , female)
# le.transform(le.classes_) - give the corresponding numerical labels [0,1]
# zip() - pairs the each original category with its assigned number
# (male , 0),(female ,1)

# display the mappings

for column, mapping in label_mappings.items():
  print(f"Mapping for {column} : {mapping}")

```

Statistic Analysis
1.Steps to perform hypothesis testing
2.Data Collection
3.preparing samples
4.To create Null and Alternate hypothesis
5.Apply the Appropriate test
6.if p-value > 0.05 Fail to reject the null hypothesis
7.if p-value < 0.05 Reject to fail the null hypothesis (accept null hypothesis)
a) It was observed that the average purchase made by the Men of the age 18-25 was 10000. Is it still the same?

```python
# one sample t-test - comparing the sample mean to a known value
# null hypothesis - mean will be equal to 10000
# alternate hypothesis - mean will not be equal to 10000
```

```python
new_data = df[(df['Age'] == 1) & (df['Gender'] == 1)]
new_data
```

```python
new_data['Purchase'].mean()
```


Machine learning models often involve random processes (e.g., splitting data, initializing weights). Setting random_state ensures that you get the same results every time you run the code.

```python
sample = new_data.sample(3600,random_state=0)
sample
```

```python
from scipy.stats import ttest_1samp
a_mean = 10000

t_stat,p_value = ttest_1samp(sample['Purchase'],a_mean)
print(t_stat,p_value)
```

```python
# As the p value is < 0.05 we are rejecting the null hypothesis and accept alternate
# mean will not be equal to 10000
```
